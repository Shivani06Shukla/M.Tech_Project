```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob

```

# Importing data :

```python
months = ['sep','oct','nov','dec','jan']
```

```python
dff = {}
for mon in months:
    
    df = pd.read_csv(f'smps_dlog_dp_{mon}_dir.csv',index_col='DateTime') 
    
    dff[mon] = df
```

```python
dff['sep']
```

## Counter Plot

```python
x=np.arange(0,288)
```

```python
#x
```

```python
y=(dff['sep'].columns)
```

```python
y = list(y)
```

```python
yf = [float(i) for i in y]
```

```python
#yf
```

```python
zn=dff['nov'].values.T
zs=dff['sep'].values.T
zo=dff['oct'].values.T
zd=dff['dec'].values.T
```

```python
zs
```

```python
zzn=zn.flatten()
plt.hist(zzn)
```

```python
zzs=zs.flatten()
zzn=zn.flatten()
zzo=zo.flatten()
zzd=zd.flatten()
```

```python
plt.figure(figsize=(10,3))

plt.subplot(1,4,1)
plt.hist(zzs)
plt.xlim(0,50000)
plt.title('sep')

plt.subplot(1,4,2)
plt.hist(zzo)
plt.xlim(0,50000)
plt.title('oct')

plt.subplot(1,4,3)
plt.hist(zzn)
plt.xlim(0,50000)
plt.title('nov')


plt.subplot(1,4,4)
plt.hist(zzd)
plt.xlim(0,50000)
plt.title('dec')
```

```python
def cnt_plot(df):
    plt.contourf(x, yf , df ,60,vmin=1000,vmax=30000,cmap='jet')
    plt.yscale("log")
    plt.xticks(ticks=[0,70,140,210,287],labels=['00','06','12','18','24'])
    plt.title(f'{mon}',fontsize=15)
    plt.colorbar()
```
