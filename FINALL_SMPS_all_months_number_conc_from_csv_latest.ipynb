# 7.2 AREA Fraction

```python
def plot_area_fc(norm_df,month):
    norm_df.plot.area(stacked=True,title=f'{month}',grid=True,fontsize=20,figsize=(4,3),legend=False)
    #plt.legend(['Nucleation','Aitkin','Accumulation'],loc='center left', bbox_to_anchor=(1.0, 0.5),fontsize=18)
    plt.xlabel("Hour of day",fontsize=15)
    #plt.ylabel("fraction PN (/cm^3)",fontsize=12)
    plt.title(f"{month}",fontsize=22)
    xtick = [0,68,137,205,274]
    xlabel = ['00','06','12','18','24']
    plt.xlim(0,274)
    plt.ylim(0,1)
    plt.ticklabel_format(axis="y", style="sci", scilimits=(0,0),useMathText=True)
    #plt.yticks([])
    plt.yticks(fontsize=15)
    plt.xticks(ticks=xtick,labels=xlabel,fontsize=15)
```

```python
for (month,file) in zip(months,files):
    df = file
    norm_df = df.div(df.sum(axis=1), axis=0);
    plot_area_fc(norm_df,month)
    
```

```python
# Monthly :

st_sep = df_st_5min.loc['2022-09']
st_oct = df_st_5min.loc['2022-10']
st_nov = df_st_5min.loc['2022-11']
st_dec = df_st_5min.loc['2022-12']
```

```python
st_sep = st_sep.reset_index()
st_oct = st_oct.reset_index()
st_nov = st_nov.reset_index()
st_dec = st_dec.reset_index()
```

```python
dir_sep = pd.DataFrame(st_sep.groupby(st_sep.DateTime.dt.time).mean()['Median (nm)'])
dir_oct = pd.DataFrame(st_oct.groupby(st_oct.DateTime.dt.time).mean()['Median (nm)'])
dir_nov = pd.DataFrame(st_nov.groupby(st_nov.DateTime.dt.time).mean()['Median (nm)'])
dir_dec = pd.DataFrame(st_dec.groupby(st_dec.DateTime.dt.time).mean()['Median (nm)'])
```

```python
dir_sep
```

```python
dir_sep = dir_sep.reset_index(drop=True).rolling(15).mean()
dir_oct = dir_oct.reset_index(drop=True).rolling(15).mean()
dir_nov = dir_nov.reset_index(drop=True).rolling(15).mean()
dir_dec = dir_dec.reset_index(drop=True).rolling(15).mean()
```

```python
plt.figure(figsize=(10,6))

plt.plot(dir_sep,label='sep')
plt.plot(dir_oct,label='oct')
plt.plot(dir_nov,label='nov')
plt.plot(dir_dec,label='dec')

plt.legend(fontsize=15)

xtick = [15,75,150,220,288]
xlabel = ['00','06','12','18','24']

plt.xticks(ticks=xtick,labels=xlabel)
plt.title("Median Diameter(nm)",fontsize=(18))
plt.ylabel("")

plt.show()
```
